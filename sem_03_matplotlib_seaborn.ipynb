```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = np.array(range(72,106))
a = np.array([4,5,2,3,4,5,6,7,8,9,1,2,4,2,3,2,3,1,8,4, 5,6,2,1,0,8,3,2,1,2,3,4,0,1])# list, series
b = np.array(range(5,72,2))
len(a)
```

```python
df = pd.DataFrame({
    'col1':a, 'col2':b
})
```

# Ломаная линия

```python
plt.plot(b) # индексы по x, значения по y
```

```python
plt.plot(b, a);
```

```python
l = np.linspace(10, 135, 34)
plt.plot(l, a, color='red', linewidth=3,linestyle='dashed', marker='*', label='Линия')
plt.legend(fontsize=10)
plt.grid(color='black', linestyle='-', linewidth=1)
plt.title('График', fontsize=22)
plt.xlabel('ось x', fontsize=15)
plt.ylabel('ось y', fontsize=15);
plt.xticks(range(5,135,10)); # разбивка значений по оси
```
